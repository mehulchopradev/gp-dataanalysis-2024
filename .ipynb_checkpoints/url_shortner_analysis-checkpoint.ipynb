```python
from json import loads
from pandas import DataFrame
```

```python
path = 'data/url-shortner/usagov_bitly_data2012-03-16-1331923249.txt'
lines = []
with open(path, mode='r') as file:
    for line in file:
        obj = loads(line)
        lines.append(obj)
```

```python
d1 = DataFrame(lines)
d1
```

```python
# top 10 timezones where the url shorterning service is used a lot!
```

```python
d1['tz'].unique()
```

```python
s1 = d1['tz']
s1[(s1 == '') | (s1.isnull())]
```

```python
s2 = s1.fillna('Missing')
s2[(s2 == '') | (s2.isnull())]
```

```python
s2[s2 == ''] = 'Missing'
s2[(s2 == '') | (s2.isnull())]
```

```python
d1['tz'] = s2
```

```python
d1.head(20)
```

```python
d2 = d1['tz'].value_counts().head(10)
d2
```

```python
d2.plot(kind='barh')
```
